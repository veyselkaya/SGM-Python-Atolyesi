```python
def factorialCalculation(sayi):
  faktoriyel = 1
  print(f"Girdiğin sayı: {sayi} ")
  if sayi ==0 or sayi ==1:
    print(f"girdiğin sayı: {sayi}. faktöriyeli ise 1 dir ")
  else:
    while sayi>1:
      faktoriyel *=sayi
      sayi -=1
    print(f"Girdiğin sayının faktöriyeli: {faktoriyel}")
    print(f"Girdiğiniz sayı: {len(str(faktoriyel))} basamaklıdır  ")
```

```python
factorialCalculation(4)
```

```python
factorialCalculation(10)
```

```python
factorialCalculation(45623)
```

```python
#Pisagor teoremi: 3-4-5, 6-8-10,.... c**2 = a**2+b**2 
#veirlen aralıkta kaç adet kenarları tamsayı olan pisagor üçgeninin olduğunu bulalım
```

```python
def pisagorBul(sayi):
  pisagor_ucgenleri= []
  for a in range(1, sayi+1):
    for b in range(1, sayi+1):
      c = (a**2+b**2)**0.5
      if c == int(c):
        pisagor_ucgenleri.append((a,b,int(c)))

  print(pisagor_ucgenleri)
  print(len(pisagor_ucgenleri))
```

```python
pisagorBul(10)
```

```python
pisagorBul(20)
```

```python
pisagorBul(1000)
```

```python
pisagorBul(10000)
```
